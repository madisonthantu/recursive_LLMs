```python
def load_json_to_dict(file_path, file_name):
    with open(os.path.join(file_path, file_name)) as json_data:
        d = json.load(json_data)
        json_data.close()
        return d
    
def load_pickle(file_path, file_name):
    print(os.path.join(file_path , file_name))
    with open(os.path.join(file_path , file_name), 'rb') as f:
        return pickle.load(f)
```

```python
root_path = '/Users/madisonthantu/Desktop/COMS_6998/Final_Project/recursive_LLMs/results/baseline'
file_paths = {
    'reddit_sharded':'reddit/sharded',
    'news_sharded':'news/sharded',
    'dialogue':'dialogue'
}
results = {}
for k, v in file_paths.items():
    results[k] = load_pickle(os.path.join(root_path, v), 'dataset_measurements.pkl')
prettyprint_dict(results)

columns = ['dataset', 'generation', 'no_total_samples', 'no_training_samples', 'no_validation_samples', 'average_doc_toks', 'average_summ_toks']
df = pd.DataFrame(columns=columns)
for k, v in results.items():
    df.loc[len(df)] = {
        'dataset':k,
        'generation':v['generation'],
        'no_total_samples':v['no_total_samples'],
        'no_training_samples':v['no_training_samples'] if 'no_training_samples' in v.keys() else v['no_train_samples'],
        'no_validation_samples':v['no_validation_samples'],
        'average_doc_toks':v['metrics']['average_doc_toks'],
        'average_summ_toks':v['metrics']['average_summ_toks'],
    }
```
